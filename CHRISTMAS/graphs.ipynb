```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
import os.path
```

```python
data_files = [file for file in os.listdir(os.getcwd()) if 'Group.dat' in file]
for f in data_files:
    data = np.loadtxt(f)
    l = "Group {0}".format(f[4])
    plt.plot(data, label=l)
    plt.legend(loc='upper right')
plt.xlabel("Node")
plt.ylabel("Flux")
plt.axvline(50, color='red', linestyle='--')
plt.title(u'{0}-group Flux, $\delta W = 10 \cdot W$'.format(len(data_files)))
plt.savefig('weird_flux.png', dpi=1000)
plt.show()
```
